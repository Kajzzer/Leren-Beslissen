# Keras

```python
import numpy as np
import pandas as pd

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD

```

```python
# get the data sets
x_train = pd.read_csv("preprocessed_train_split_feat.csv")
x_test = pd.read_csv("preprocessed_validation_split_feat.csv")

# get the labels
y_train = np.asarray(list(pd.read_csv("train_split_label.csv").Label))
y_test = np.asarray(list(pd.read_csv("validation_split_label.csv").Label))

print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
def numeric_labels(data):
    labels = []
    for label in data:
        if label == 'sitting':
            labels.append(0.1)
        elif label == 'lift':
            labels.append(0.2)
        elif label =='towlift':
            labels.append(0.3)
        elif label =='standing':
            labels.append(0.4)
        elif label =='lying':
            labels.append(0.5)
        elif label =='snowboarding':
            labels.append(0.6)
    return np.asarray(labels).astype(float)     
```

```python
y_train = numeric_labels(y_train)
y_test = numeric_labels(y_test)

print(y_train.shape, y_test.shape)
print(y_test)
```

```python
model = Sequential()
# Dense(64) is a fully-connected layer with 64 hidden units.
# in the first layer, you must specify the expected input data shape:
# here, 20-dimensional vectors.
model.add(Dense(32, activation='relu', input_dim=16))
model.add(Dropout(0.5))
# model.add(Dense(32, activation='relu'))
# model.add(Dropout(0.5))
model.add(Dense(6, activation='softmax'))

sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='sparse_categorical_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])

model.fit(x_train, y_train,
          epochs=20,
          batch_size=8)
score = model.evaluate(x_test, y_test, batch_size=8)
pred = model.predict(x_test)
```
